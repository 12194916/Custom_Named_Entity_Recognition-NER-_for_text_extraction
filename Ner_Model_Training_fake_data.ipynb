```python
df = pd.DataFrame(text)
df.head()
```

```python
order_num_pattern = r'№\s*(\d+)'
date_pattern = r'(\d{2}\.\d{2}\.\d{4})'
df['order_number'] = df['text'].str.extract(order_num_pattern)
df['date'] = df['text'].str.extract(date_pattern)
df
```

```python
df.isnull().sum()
```

```python
dataset = [(row['text'], (row['order_number'], row['date'])) for index, row in df.iterrows()]
dataset
```

```python
TRAIN_DATA = []
for text, entity_text in dataset:
    entities = []
    start = 0
    for token in entity_text:
        start = text.find(token, start)
        end = start + len(token)
        entities.append((start, end, "ENTITY"))
        start = end
    TRAIN_DATA.append({"text": text, "entities": entities})
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
nlp.pipe_names
```

```python
import spacy
from spacy.training.example import Example
import random

# create a blank English model and add the named entity recognizer
nlp = spacy.blank('en')
ner = nlp.add_pipe('ner')

# add the labels
for annotation in TRAIN_DATA:
    for entity in annotation['entities']:
        ner.add_label(entity[2])

# train the model
nlp.begin_training()
for epoch in range(10):
    random.shuffle(TRAIN_DATA)
    for example in TRAIN_DATA:
        # create a spaCy example
        doc = nlp.make_doc(example['text'])
        entities = example['entities']
        example = Example.from_dict(doc, {'entities': entities})

        # update the model with the example
        nlp.update([example], drop=0.5)
```

```python
# test the model
doc = nlp(' на основание дог комисси № 7245 от 15.01.2021 года оплата  за алюмин профиль,')
```
